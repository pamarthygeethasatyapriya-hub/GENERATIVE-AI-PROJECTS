````python
import streamlit as st
import wikipedia
import random

st.title("🤖 AI Question Generator (Accurate Version)")

topic = st.text_input("Enter Topic")

# Clean text into sentences
def get_sentences(text):
    return [s.strip() for s in text.split('.') if len(s) > 20]

# Generate MCQ from sentence
def create_mcq(sentence, topic):
    words = sentence.split()

    # pick a keyword from sentence
    key = random.choice([w for w in words if len(w) > 6])

    question = sentence.replace(key, "_____")

    # options
    options = [key]

    distractors = random.sample(words, min(3, len(words)))
    options.extend(distractors)

    options = list(set(options))
    random.shuffle(options)

    return question, options

def generate_questions(topic):
    try:
        summary = wikipedia.summary(topic, sentences=4)
    except:
        return None, None, None, "Topic not found"

    sentences = get_sentences(summary)

    mcqs = []
    for s in sentences[:3]:
        q, opts = create_mcq(s, topic)
        mcqs.append((q, opts))

    short_qs = [
        f"Define {topic}.",
        f"Explain the importance of {topic}."
    ]

    long_qs = [
        f"Explain {topic} in detail with examples.",
        f"Discuss applications of {topic}."
    ]

    return mcqs, short_qs, long_qs, summary


if st.button("Generate Questions"):
    if topic:
        mcqs, shorts, longs, summary = generate_questions(topic)

        if mcqs is None:
            st.error(summary)
        else:
            st.subheader("📘 Topic Summary")
            st.write(summary)

            st.subheader("Section A: MCQs")
            for i, (q, opts) in enumerate(mcqs, 1):
                st.write(f"{i}. {q}")
                for opt in opts:
                    st.write(f"- {opt}")

            st.subheader("Section B: Short Questions")
            for i, q in enumerate(shorts, 1):
                st.write(f"{i}. {q}")

            st.subheader("Section C: Long Questions")
            for i, q in enumerate(longs, 1):
                st.write(f"{i}. {q}")
    else:
        st.warning("Please enter a topic!")
````
